# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
print('Введите числа.')
a=input()
s=a.split()
mx=0
sm=0
k=0
for i in s:
  if mx<int(i):
    mx=int(i)
  k+=1
  sm+=int(i)
print('Сумма чисел:', sm, 'Среднее арифм.:', sm/k, 'Максимальное значение:', mx)
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random
print('Давай сыграем в игру Камень, ножницы, бумага! Играем до 3 побед!')
win_player=0
win_computer=0
while win_player<3 and win_computer<3:
  a=input()
  s=['Ножницы', 'Камень', 'Бумага']
  c=random.choice(s)
  print('Компьютер:', c)
  print('Игрок:', a)
  if a==c:
    print('Победа: Ничья')
  elif a==s[0]:
    if c==s[1]:
      print('Победа: Компьютер')
      win_computer+=1
    else:
      print('Победа: Игрок')
      win_player+=1
  elif a==s[1]:
    if c==s[2]:
      print('Победа: Компьютер')
      win_computer+=1
    else:
      print('Победа: Игрок')
      win_player+=1
  elif a==s[2]:
    if c==s[0]:
      print('Победа: Компьютер')
      win_computer+=1
    else:
      print('Победа: Игрок')
      win_player+=1
if win_player>win_computer:
  print('Победил игрок')
else:
  print('Победил компьютер')
````

# **Задача 5.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру (игра должна иметь привлекательный интерфейс, соотвесвующий классической висилице)

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
import random
print('Давай сыграем в игру "Виселица"! Чур я загадываю слово! Если правильно угадываешь слово, твои очки за новые буквы удваиваются!')
s=['мебель', 'лестница', 'дельфин', 'медведь', 'клавиатура', 'солнце']
c=random.choice(s)
s2=[]
s3=[]
for i in range(1, len(c)-1):
  s2.append(c[i])
  s3.append('_')
print('Я загадал слово!')
print(c[0], ' '.join(s3), c[-1])
x=6
g=0
j=0
while x>0 and g<len(s2):
  print('Угадывай буквы')
  b=input()
  if b==c:
    j+=2*s3.count('_')
    break
  elif s2.count(b)==0:
    x-=1
    print('Ни одной!')
    print('У тебя осталось', x, 'попыток. Твои очки:', j)
    print(c[0], ' '.join(s3), c[-1])
  elif s2.count(b)==1:
    ind=s2.index(b)
    s3.insert(ind, s2[ind])
    s3.pop(ind+1)
    print('У тебя осталось', x, 'попыток. Твои очки:', j)
    print(c[0], ' '.join(s3), c[-1])
    g+=1
    j+=1
  else:
    f=s2.count(b)
    j+=f
    while f>0:
      f-=1
      ind=s2.index(b)
      s3.insert(ind, s2[ind])
      s3.pop(ind+1)
      s2.insert(ind, '_')
      s2.pop(ind+1)
    print('У тебя осталось', x, 'попыток. Твои очки:', j)
    print(c[0], ' '.join(s3), c[-1])
    g+=1
if x==0:
  print('Твои попытки закончились! Ты проиграл. Я загадал слово:', c)
else:
  print('Правильно!', c, '!', 'Ты победил! Твои очки:', j)
````
