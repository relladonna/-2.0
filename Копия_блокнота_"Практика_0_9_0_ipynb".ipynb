Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unique_elements(s1, s2):
  s = []
  for i in s1:
    if i in s2:
      s2.pop(s2.index(i))
    else:
      s.append(i)
  for v in s2:
    s.append(v)
  return s


a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
print(unique_elements(a, b))
````
